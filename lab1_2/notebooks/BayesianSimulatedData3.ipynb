# BayesianSimulatedData3

```python
import numpy as np
import matplotlib.pyplot as plt
from tests.generate_dataset import generate_dataset_scheme1, generate_dataset_scheme2
from tests.train_test_split import train_test_split
from src.QDAClassifier import QDAClassifier
from src.NBClassifier import NBClassifier
from src.LDAClassifier import LDAClassifier
```

```python
X, y = generate_dataset_scheme1(a=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, 0.7)
```

```python
virtual_data = np.arange(-3,5, 0.05)
virtual_data_points = np.array([[x,y] for x in virtual_data for y in virtual_data])
```

## LDA - scheme 1

```python
lda = LDAClassifier()
lda.fit(X_train, y_train)
```

```python
proba = lda.predict_proba(virtual_data_points)
decision_boundary = virtual_data_points[(proba > 0.48) & (proba < 0.52)]
```

```python
fig = plt.subplot( )
fig.scatter(decision_boundary[:, 0], decision_boundary[:, 1], s=1, c='red')
fig.scatter(X_train[y_train==0, 0], X_train[y_train==0, 1], label='Class 0', alpha=0.5)
fig.scatter(X_train[y_train==1, 0], X_train[y_train==1, 1], label='Class 1', alpha=0.5)
fig.set_xlabel('Feature 1')
fig.set_ylabel('Feature 2')
fig.axis('equal')
fig.grid()
```
